# BAND NAME GENERATOR

```python
# Band Name Generator

import random

# Predefined lists of words for band names
adjectives = ["Rocking", "Silent", "Electric", "Funky", "Crazy", "Golden", "Burning", "Epic"]
nouns = ["Wolves", "Dragons", "Eagles", "Beats", "Stars", "Pirates", "Rebels", "Titans"]

def generate_band_name(theme="default"):
    """Generates a random band name based on the selected theme."""
    if theme == "rock":
        adjectives = ["Loud", "Heavy", "Thunderous", "Fiery", "Raging", "Vibrant"]
        nouns = ["Storm", "Riffs", "Strikers", "Phoenix", "Blaze", "Demons"]
    elif theme == "fantasy":
        adjectives = ["Mystical", "Enchanted", "Magical", "Ancient", "Dark"]
        nouns = ["Elves", "Kingdom", "Shadows", "Legends", "Crystals"]
    else:
        adjectives = ["Rocking", "Silent", "Electric", "Funky", "Crazy", "Golden", "Burning", "Epic"]
        nouns = ["Wolves", "Dragons", "Eagles", "Beats", "Stars", "Pirates", "Rebels", "Titans"]

    adj = random.choice(adjectives)
    noun = random.choice(nouns)
    return f"{adj} {noun}"

# Main function
def main():
    print("Welcome to the Band Name Generator!")
    print("Choose a theme for your band name: ")
    print("1. Default")
    print("2. Rock")
    print("3. Fantasy")

    theme_choice = input("Enter your choice (1/2/3): ").strip()
    theme_map = {"1": "default", "2": "rock", "3": "fantasy"}
    theme = theme_map.get(theme_choice, "default")

    while True:
        band_name = generate_band_name(theme)
        print(f"Your generated band name: {band_name}")

        save_choice = input("Do you want to save this name? (yes/no): ").strip().lower()
        if save_choice == "yes":
            with open("favorite_band_names.txt", "a") as file:
                file.write(band_name + "\n")
            print(f"Band name '{band_name}' saved!")

        another = input("Generate another name? (yes/no): ").strip().lower()
        if another != "yes":
            break

    print("Thank you for using the Band Name Generator!")

if __name__ == "__main__":
    main()
```

# ROCK PAPER SCISSORS

```python
import random
from IPython.display import clear_output

def display_score(player_score, computer_score):
    """
    Displays the current score.
    """
    print(f"Score: You {player_score} - {computer_score} Computer\n")

def play_round():
    """
    Plays a single round of Rock-Paper-Scissors.
    """
    # List of valid choices
    choices = ["rock", "paper", "scissors"]

    # Get player choice
    player_choice = input("Choose Rock, Paper, or Scissors: ").lower()
    if player_choice not in choices:
        print("\nInvalid choice! Please select 'rock', 'paper', or 'scissors'.")
        return None, None

    # Get computer choice
    computer_choice = random.choice(choices)

    # Determine the winner
    if player_choice == computer_choice:
        return "draw", computer_choice
    elif (player_choice == "rock" and computer_choice == "scissors") or \
         (player_choice == "scissors" and computer_choice == "paper") or \
         (player_choice == "paper" and computer_choice == "rock"):
        return "player", computer_choice
    else:
        return "computer", computer_choice

def rock_paper_scissors():
    """
    Main function to play the Rock-Paper-Scissors game.
    """
    print("Welcome to Rock-Paper-Scissors!")
    print("Type 'exit' to quit the game at any time.\n")

    player_score = 0
    computer_score = 0

    while True:
        display_score(player_score, computer_score)

        # Play a round
        result, computer_choice = play_round()

        # Handle game results
        if result == "draw":
            print("\nIt's a draw!")
        elif result == "player":
            print("\nYou win this round!")
            player_score += 1
        elif result == "computer":
            print("\nYou lose this round!")
            computer_score += 1
        else:
            # Invalid choice or exit
            player_input = input("\nDo you want to continue? Type 'yes' to play or 'exit' to quit: ").lower()
            if player_input == "exit":
                break
            else:
                continue

        # Ask player if they want to play another round
        play_again = input("\nPlay another round? Type 'yes' to continue or 'exit' to quit: ").lower()
        if play_again == "exit":
            break

    # Final score display
    clear_output(wait=True)
    display_score(player_score, computer_score)
    print("Game Over! Thanks for playing!")
    if player_score > computer_score:
        print("ðŸŽ‰ You are the overall winner! ðŸŽ‰")
    elif player_score < computer_score:
        print("ðŸ˜¢ The computer wins this time! ðŸ˜¢")
    else:
        print("ðŸ¤ It's a draw overall! ðŸ¤")

# Start the game
rock_paper_scissors()
```

# PASSWORD GENERATOR

```python
import random
import string
import re

def generate_password(length=12):
    """
    Generates a random password with the specified length.
    The password will include uppercase, lowercase, digits, and special characters.
    """
    # Define possible characters
    uppercase = string.ascii_uppercase
    lowercase = string.ascii_lowercase
    digits = string.digits
    special_characters = string.punctuation

    # Combine all characters
    all_characters = uppercase + lowercase + digits + special_characters

    # Randomly select characters for the password
    password = ''.join(random.choice(all_characters) for i in range(length))

    return password

def check_password_strength(password):
    """
    Checks the strength of the password and returns a strength rating.
    """
    strength = 0
    # Check length
    if len(password) >= 12:
        strength += 1
    # Check for uppercase, lowercase, digits, and special characters
    if re.search(r'[A-Z]', password):
        strength += 1
    if re.search(r'[a-z]', password):
        strength += 1
    if re.search(r'[0-9]', password):
        strength += 1
    if re.search(r'[!@#$%^&*(),.?":{}|<>]', password):
        strength += 1
    # Check for common dictionary words (basic check)
    if re.search(r'\b(?:password|1234|qwerty|abc|letmein)\b', password, re.IGNORECASE):
        strength -= 1  # Weak if it contains common words

    if strength == 5:
        return "Very Strong"
    elif strength == 4:
        return "Strong"
    elif strength == 3:
        return "Moderate"
    elif strength == 2:
        return "Weak"
    else:
        return "Very Weak"

# Input for password length
length = int(input("Enter the desired password length (default is 12): ") or 12)

# Generate the password
password = generate_password(length)

# Check password strength
strength = check_password_strength(password)

# Display the generated password and its strength
print(f"\nGenerated Password: {password}")
print(f"Password Strength: {strength}")
```

# ESCAPING THE MAZE

```python
import random
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import time

# Constants for maze generation
DIFFICULTY = {'easy': 7, 'medium': 10, 'hard': 15}  # Maze size (easy = 7x7, medium = 10x10, hard = 15x15)
WALL = 1
PATH = 0

# Directions for movement
MOVES = [(0, 1), (1, 0), (0, -1), (-1, 0)]  # Right, Down, Left, Up

def generate_maze(size):
    """
    Generates a random maze using depth-first search (DFS).
    """
    maze = np.ones((size, size), dtype=int)  # Initialize maze with walls (1)
    stack = [(1, 1)]  # Start position (1, 1), typically at top-left corner
    maze[1][1] = 0  # Mark the start position as a path (0)

    while stack:
        x, y = stack[-1]
        neighbors = []

        for dx, dy in MOVES:
            nx, ny = x + dx * 2, y + dy * 2  # Skip 1 block to create a path
            if 0 < nx < size-1 and 0 < ny < size-1 and maze[nx][ny] == 1:
                neighbors.append((nx, ny, dx, dy))

        if neighbors:
            nx, ny, dx, dy = random.choice(neighbors)
            maze[nx][ny] = 0  # Mark new cell as path
            maze[x + dx][y + dy] = 0  # Mark connecting path
            stack.append((nx, ny))
        else:
            stack.pop()

    # Mark the exit
    maze[size-2][size-2] = 0  # Exit position near bottom-right corner
    return maze

def display_maze(maze, player_x, player_y):
    """
    Displays the maze using matplotlib with color.
    """
    cmap = mcolors.ListedColormap(['white', 'black', 'green', 'red'])
    display_maze = maze.copy()

    # Mark the player's position as green
    display_maze[player_x][player_y] = 2
    # Mark the exit position as red
    display_maze[maze.shape[0]-2][maze.shape[1]-2] = 3

    plt.imshow(display_maze, cmap=cmap)
    plt.axis('off')
    plt.show()

def move_player(maze, x, y, direction):
    """
    Moves the player based on the direction and updates position.
    """
    dx, dy = MOVES[direction]
    new_x, new_y = x + dx, y + dy

    if maze[new_x][new_y] == 0:  # Check if it's a valid path
        return new_x, new_y
    return x, y  # No movement if blocked

def choose_difficulty():
    """
    Prompts the player to choose a difficulty level.
    """
    print("Choose difficulty level: easy, medium, or hard.")
    difficulty = input("Enter difficulty: ").lower()
    while difficulty not in DIFFICULTY:
        print("Invalid difficulty level! Please choose from 'easy', 'medium', or 'hard'.")
        difficulty = input("Enter difficulty: ").lower()
    return DIFFICULTY[difficulty]

def play_game():
    """
    Main function to play the Escape the Maze game.
    """
    # Choose the difficulty and generate the maze
    maze_size = choose_difficulty()
    maze = generate_maze(maze_size)

    player_x, player_y = 1, 1  # Starting position
    exit_x, exit_y = maze_size - 2, maze_size - 2  # Exit position

    maze[exit_x][exit_y] = 0  # Ensure the exit is a valid path

    # Timer start
    start_time = time.time()

    print(f"\nWelcome to the Maze! Reach the exit to win. Difficulty: {maze_size}x{maze_size}")
    print("Use WASD to move (W = Up, A = Left, S = Down, D = Right).")
    print(f"You are at position (1, 1), and the exit is at ({exit_x}, {exit_y}).")

    while True:
        display_maze(maze, player_x, player_y)

        move = input("Enter your move (W/A/S/D): ").lower()

        if move == 'w':  # Up
            player_x, player_y = move_player(maze, player_x, player_y, 3)
        elif move == 'a':  # Left
            player_x, player_y = move_player(maze, player_x, player_y, 2)
        elif move == 's':  # Down
            player_x, player_y = move_player(maze, player_x, player_y, 1)
        elif move == 'd':  # Right
            player_x, player_y = move_player(maze, player_x, player_y, 0)
        else:
            print("Invalid move! Please use W, A, S, or D.")
            continue

        # Check if player has reached the exit
        if player_x == exit_x and player_y == exit_y:
            end_time = time.time()
            elapsed_time = round(end_time - start_time, 2)
            print(f"Congratulations! You've escaped the maze in {elapsed_time} seconds!")
            break

# Run the game
play_game()
```

```python
import random

# List of words for the game
WORD_BANK = [
    "python", "hangman", "developer", "algorithm", "function",
    "variable", "iteration", "string", "dictionary", "inheritance"
]

def choose_word():
    """
    Randomly selects a word from the WORD_BANK.
    """
    return random.choice(WORD_BANK)

def display_progress(guessed_word, guessed_letters, attempts):
    """
    Displays the current progress of the game.
    """
    print("\nCurrent Word: ", " ".join(guessed_word))
    print(f"Guessed Letters: {', '.join(sorted(guessed_letters)) if guessed_letters else 'None'}")
    print(f"Remaining Attempts: {attempts}")

def play_hangman():
    """
    Main function to play the Hangman game.
    """
    word = choose_word()  # Select a random word
    guessed_word = ["_"] * len(word)  # Placeholder for guessed letters
    attempts = 6  # Number of wrong attempts allowed
    guessed_letters = set()  # Keep track of guessed letters

    print("Welcome to Hangman!")
    print("Try to guess the word letter by letter.")
    print("You have 6 attempts to figure out the word.\n")

    while attempts > 0 and "_" in guessed_word:
        # Display the game progress
        display_progress(guessed_word, guessed_letters, attempts)

        # Get player's guess
        guess = input("Enter a letter: ").lower()

        if len(guess) != 1 or not guess.isalpha():
            print("Invalid input! Please enter a single letter.")
            continue

        if guess in guessed_letters:
            print(f"You've already guessed '{guess}'. Try a different letter.")
            continue

        guessed_letters.add(guess)

        if guess in word:
            print(f"Good job! '{guess}' is in the word.")
            for idx, letter in enumerate(word):
                if letter == guess:
                    guessed_word[idx] = guess
        else:
            print(f"Oops! '{guess}' is not in the word.")
            attempts -= 1

    # Game over
    if "_" not in guessed_word:
        print("\nðŸŽ‰ Congratulations! You guessed the word! ðŸŽ‰")
        print(f"The word was: {word}")
    else:
        print("\nðŸ˜¢ You ran out of attempts! Better luck next time! ðŸ˜¢")
        print(f"The word was: {word}")

# Start the game
play_hangman()
```

```python
import random

# List of words for the game
WORD_BANK = [
    "python", "hangman", "developer", "algorithm", "function", "variable",
    "iteration", "string", "dictionary", "optimization", "exception", "important"
]

# Hangman stages (ASCII art)
HANGMAN_STAGES = [
    '''
      +---+
      |   |
          |
          |
          |
          |
    =========
    ''',
    '''
      +---+
      |   |
      O   |
          |
          |
          |
    =========
    ''',
    '''
      +---+
      |   |
      O   |
      |   |
          |
          |
    =========
    ''',
    '''
      +---+
      |   |
      O   |
     /|   |
          |
          |
    =========
    ''',
    '''
      +---+
      |   |
      O   |
     /|\\  |
          |
          |
    =========
    ''',
    '''
      +---+
      |   |
      O   |
     /|\\  |
     /    |
          |
    =========
    ''',
    '''
      +---+
      |   |
      O   |
     /|\\  |
     / \\  |
          |
    =========
    '''
]

def choose_word():
    """
    Randomly selects a word from the WORD_BANK.
    """
    return random.choice(WORD_BANK)

def display_hangman(attempts):
    """
    Displays the current state of the hangman.
    """
    print(HANGMAN_STAGES[len(HANGMAN_STAGES) - 1 - attempts])

def play_hangman():
    """
    Main function to play the Hangman game.
    """
    word = choose_word()
    guessed_word = ["_"] * len(word)  # Placeholder for guessed letters
    attempts = 6  # Number of wrong attempts allowed
    guessed_letters = set()  # Keep track of guessed letters

    print("Welcome to Hangman!")
    print("Try to guess the word before the hangman is fully drawn.")
    print("You have 6 incorrect guesses.\n")

    while attempts > 0 and "_" in guessed_word:
        # Display current game state
        display_hangman(attempts)
        print("\nWord: ", " ".join(guessed_word))
        print(f"Guessed letters: {', '.join(sorted(guessed_letters)) if guessed_letters else 'None'}")
        print(f"Remaining attempts: {attempts}")

        # Get player's guess
        guess = input("\nEnter a letter: ").lower()

        if len(guess) != 1 or not guess.isalpha():
            print("\nInvalid input! Please enter a single letter.")
            continue

        if guess in guessed_letters:
            print(f"\nYou've already guessed '{guess}'. Try another letter.")
            continue

        guessed_letters.add(guess)

        if guess in word:
            print(f"\nGood job! '{guess}' is in the word.")
            for idx, letter in enumerate(word):
                if letter == guess:
                    guessed_word[idx] = guess
        else:
            print(f"\nOops! '{guess}' is not in the word.")
            attempts -= 1

    # Game over
    if "_" not in guessed_word:
        print("\nðŸŽ‰ Congratulations! You've guessed the word! ðŸŽ‰")
        print(f"The word was: {word}")
    else:
        display_hangman(0)
        print("\nðŸ˜¢ You ran out of attempts! The hangman is complete. ðŸ˜¢")
        print(f"The word was: {word}")

# Start the game
play_hangman()
```

# CALCULATOR

```python
import math

def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    if b == 0:
        return "Error! Division by zero."
    return a / b

def power(a, b):
    return a ** b

def sqrt(a):
    if a < 0:
        return "Error! Cannot calculate the square root of a negative number."
    return math.sqrt(a)

def log(a, base):
    if a <= 0:
        return "Error! Logarithm undefined for non-positive values."
    return math.log(a, base)

def sin(a):
    return math.sin(math.radians(a))

def cos(a):
    return math.cos(math.radians(a))

def tan(a):
    try:
        return math.tan(math.radians(a))
    except:
        return "Error! Undefined tan value."

def display_history(history):
    if not history:
        print("No calculations performed yet.")
    else:
        print("\nCalculation History:")
        for index, entry in enumerate(history, start=1):
            print(f"{index}. {entry}")
    print("\n")

def calculator():
    print("Welcome to the Enhanced Python Calculator!")
    print("Operations available:")
    print("1. Addition (+)")
    print("2. Subtraction (-)")
    print("3. Multiplication (*)")
    print("4. Division (/)")
    print("5. Power (^)")
    print("6. Square Root (âˆš)")
    print("7. Logarithm (log)")
    print("8. Sine (sin)")
    print("9. Cosine (cos)")
    print("10. Tangent (tan)")
    print("Type 'history' to view calculation history or 'exit' to quit.\n")

    history = []  # Store calculation history

    while True:
        operation = input("Enter the operation (+, -, *, /, ^, âˆš, log, sin, cos, tan, history, exit): ").strip().lower()

        if operation == "exit":
            print("Thank you for using the calculator. Goodbye!")
            break

        if operation == "history":
            display_history(history)
            continue

        if operation not in ["+", "-", "*", "/", "^", "âˆš", "log", "sin", "cos", "tan"]:
            print("Invalid operation. Please try again.")
            continue

        if operation == "âˆš":
            try:
                num = float(input("Enter the number: "))
                result = sqrt(num)
                history.append(f"âˆš({num}) = {result}")
            except ValueError:
                print("Invalid input! Please enter a valid number.")
                continue

        elif operation in ["sin", "cos", "tan"]:
            try:
                angle = float(input("Enter the angle in degrees: "))
                if operation == "sin":
                    result = sin(angle)
                    history.append(f"sin({angle}) = {result}")
                elif operation == "cos":
                    result = cos(angle)
                    history.append(f"cos({angle}) = {result}")
                elif operation == "tan":
                    result = tan(angle)
                    history.append(f"tan({angle}) = {result}")
            except ValueError:
                print("Invalid input! Please enter a valid number.")
                continue

        elif operation == "log":
            try:
                num = float(input("Enter the number: "))
                base = float(input("Enter the base: "))
                result = log(num, base)
                history.append(f"log_{base}({num}) = {result}")
            except ValueError:
                print("Invalid input! Please enter valid numbers.")
                continue

        else:
            try:
                num1 = float(input("Enter the first number: "))
                num2 = float(input("Enter the second number: "))

                if operation == "+":
                    result = add(num1, num2)
                    history.append(f"{num1} + {num2} = {result}")
                elif operation == "-":
                    result = subtract(num1, num2)
                    history.append(f"{num1} - {num2} = {result}")
                elif operation == "*":
                    result = multiply(num1, num2)
                    history.append(f"{num1} * {num2} = {result}")
                elif operation == "/":
                    result = divide(num1, num2)
                    history.append(f"{num1} / {num2} = {result}")
                elif operation == "^":
                    result = power(num1, num2)
                    history.append(f"{num1} ^ {num2} = {result}")
            except ValueError:
                print("Invalid input! Please enter valid numbers.")
                continue

        print(f"The result is: {result}\n")

# Run the enhanced calculator
calculator()
```

# BLACKJACK PROJECT

```python
import random

def deal_card():
    """
    Deals a single card randomly from the deck.
    """
    cards = [2, 3, 4, 5, 6, 7, 8, 9, 10, 10, 10, 10, 11]  # 10 for face cards, 11 for Ace
    return random.choice(cards)

def calculate_score(cards):
    """
    Calculates the total score for a given hand.
    Handles Ace being 1 or 11 dynamically.
    """
    score = sum(cards)
    if 11 in cards and score > 21:
        cards.remove(11)
        cards.append(1)  # Convert Ace from 11 to 1
    return score

def display_cards(player_cards, dealer_cards, hide_dealer_card=True):
    """
    Displays the player's and dealer's cards.
    The dealer's first card is hidden until the end.
    """
    print("\nYour cards:", player_cards, f"Score: {calculate_score(player_cards)}")
    if hide_dealer_card:
        print("Dealer's cards: [?,", *dealer_cards[1:], "]")
    else:
        print("Dealer's cards:", dealer_cards, f"Score: {calculate_score(dealer_cards)}")

def blackjack_game(balance):
    """
    Main function to play a round of Blackjack with betting.
    """
    print("\nWelcome to Blackjack!")
    print(f"Your current balance: ${balance}")

    # Place bet
    while True:
        try:
            bet = int(input(f"Place your bet (1 - {balance}): $"))
            if 1 <= bet <= balance:
                break
            else:
                print(f"Invalid bet! Enter a value between 1 and {balance}.")
        except ValueError:
            print("Invalid input! Please enter a numeric value.")

    player_cards = [deal_card(), deal_card()]
    dealer_cards = [deal_card(), deal_card()]
    game_over = False

    while not game_over:
        display_cards(player_cards, dealer_cards)

        # Check if player or dealer has a Blackjack
        if calculate_score(player_cards) == 21 or calculate_score(dealer_cards) == 21:
            game_over = True
            break

        # Ask player to hit or stand
        choice = input("\nType 'hit' to draw another card, or 'stand' to hold your cards: ").lower()
        if choice == "hit":
            player_cards.append(deal_card())
            if calculate_score(player_cards) > 21:
                game_over = True
        elif choice == "stand":
            game_over = True
        else:
            print("Invalid choice! Please type 'hit' or 'stand'.")

    # Dealer's turn
    while calculate_score(dealer_cards) < 17:
        dealer_cards.append(deal_card())

    # Final results
    display_cards(player_cards, dealer_cards, hide_dealer_card=False)

    player_score = calculate_score(player_cards)
    dealer_score = calculate_score(dealer_cards)

    print("\nGame Over!")
    if player_score > 21:
        print("ðŸ˜¢ You busted! Dealer wins.")
        return balance - bet
    elif dealer_score > 21 or player_score > dealer_score:
        print("ðŸŽ‰ You win!")
        return balance + bet
    elif player_score < dealer_score:
        print("ðŸ˜¢ Dealer wins!")
        return balance - bet
    else:
        print("ðŸ¤ It's a draw!")
        return balance

def blackjack():
    """
    Main function to handle multiple rounds and balance tracking.
    """
    balance = 100  # Starting balance
    print("Welcome to Enhanced Blackjack!")
    print("Start with $100. Win as much as you can!")

    while balance > 0:
        balance = blackjack_game(balance)
        print(f"\nYour current balance: ${balance}")

        if balance <= 0:
            print("ðŸ˜¢ You've run out of money! Game over.")
            break

        # Efficient replay prompt
        replay = input("Do you want to play another round? (yes/no): ").strip().lower()
        if replay != "yes":
            print("Thanks for playing! Come back soon!")
            break

# Start the enhanced Blackjack game
blackjack()


```

# THE NUMBER GUESSING GAME

```python
import random
import time

# Initialize leaderboards
leaderboard = []

def choose_difficulty():
    """
    Let the user choose a difficulty level.
    """
    print("\nChoose your difficulty level:")
    print("1. Easy (1 to 50)")
    print("2. Medium (1 to 100)")
    print("3. Hard (1 to 200)")

    while True:
        choice = input("Enter 1, 2, or 3: ")
        if choice == "1":
            return 50
        elif choice == "2":
            return 100
        elif choice == "3":
            return 200
        else:
            print("Invalid choice. Please choose 1, 2, or 3.")

def give_hint(number, guess):
    """
    Provide a hint based on whether the number is odd or even.
    """
    if number % 2 == 0:
        print("Hint: The number is even!")
    else:
        print("Hint: The number is odd!")

def play_game():
    """
    Main function to play the number guessing game with enhancements.
    """
    global leaderboard

    print("Welcome to the Enhanced Number Guessing Game!")

    # Choose difficulty level
    upper_bound = choose_difficulty()
    lower_bound = 1
    number_to_guess = random.randint(lower_bound, upper_bound)

    attempts = 0
    guessed = False
    score = 0
    start_time = time.time()

    print(f"Guess a number between {lower_bound} and {upper_bound}.")

    # Game loop
    while not guessed:
        try:
            guess = int(input("Enter your guess: "))
        except ValueError:
            print("Invalid input! Please enter an integer.")
            continue

        attempts += 1

        if guess < number_to_guess:
            print("Too low! Try again.")
        elif guess > number_to_guess:
            print("Too high! Try again.")
        else:
            guessed = True
            end_time = time.time()
            elapsed_time = end_time - start_time
            score = max(100 - int(elapsed_time), 10)  # Score decreases with time
            print(f"Congratulations! You guessed the number in {attempts} attempts.")
            print(f"Your score: {score}")

            # Save the score in the leaderboard
            leaderboard.append(score)
            leaderboard = sorted(leaderboard, reverse=True)[:5]  # Keep top 5 scores
            print(f"\nTop scores: {leaderboard}")

        # Give hint after 3 wrong guesses
        if attempts > 3:
            give_hint(number_to_guess, guess)

    # Ask if the player wants to play again
    play_again = input("\nDo you want to play again? (yes/no): ").lower()
    if play_again == "yes":
        play_game()  # Restart the game
    else:
        print("Thanks for playing! Goodbye.")

# Start the game
play_game()
```
